```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

```python
df = pd.read_csv("Online_Retail.csv", encoding='ISO-8859-1')
display(df.head())
df.info()
```

```python
df.isnull().sum()
```

```python
# Drop rows with missing CustomerID and Description
df.dropna(subset=['CustomerID'], inplace=True)
df.dropna(subset=['Description'], inplace=True)

# Remove cancelled orders (InvoiceNo starting with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Remove rows with non-positive Quantity or UnitPrice
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Create TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

print(f"DataFrame shape after cleaning: {df.shape}")
df.isnull().sum()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(df.columns.tolist())
df.info()
```

```python
snapshot_date = df['InvoiceDate'].max()

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalPrice': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm = rfm[rfm['Monetary'] > 0]
print(f"RFM shape: {rfm.shape}")
display(rfm.head())
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(rfm.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of RFM Features')
plt.show()
```

```python
# --- RFM Score-Based Segmentation ---
# Score each R, F, M on a scale of 1 to 3 using quantile-based binning

# Recency: lower is better → reverse scoring
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=3, labels=[3, 2, 1]).astype(int)

# Frequency: higher is better
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=3, labels=[1, 2, 3]).astype(int)

# Monetary: higher is better
rfm['M_Score'] = pd.qcut(rfm['Monetary'], q=3, labels=[1, 2, 3]).astype(int)

# Total RFM Score (max=9, min=3)
rfm['RFM_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']

print("RFM Score distribution:")
print(rfm['RFM_Score'].value_counts().sort_index())
display(rfm.head(10))
```

```python
# Assign segments based on RFM Score thresholds
# Score 7-9  → High Value
# Score 4-6  → Medium Value
# Score 3    → Low Value

def assign_segment(score):
    if score >= 7:
        return 'High Value'
    elif score >= 4:
        return 'Medium Value'
    else:
        return 'Low Value'

rfm['Segment'] = rfm['RFM_Score'].apply(assign_segment)

print("Segment distribution:")
print(rfm['Segment'].value_counts())
display(rfm.head())
```

```python
# Profile of each RFM segment
segment_profiles = rfm.groupby('Segment').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'RFM_Score': 'mean'
}).round(2)
display(segment_profiles)
```

```python
# Visualize segment counts
plt.figure(figsize=(8, 5))
order = ['High Value', 'Medium Value', 'Low Value']
counts = rfm['Segment'].value_counts().reindex(order)
colors = ['#2ecc71', '#f39c12', '#e74c3c']
plt.bar(counts.index, counts.values, color=colors, edgecolor='black')
plt.title('Customer Segment Distribution (RFM Score-Based)')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
for i, v in enumerate(counts.values):
    plt.text(i, v + 5, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Monetary vs Frequency coloured by Segment
plt.figure(figsize=(10, 7))
palette = {'High Value': '#2ecc71', 'Medium Value': '#f39c12', 'Low Value': '#e74c3c'}
for seg, color in palette.items():
    subset = rfm[rfm['Segment'] == seg]
    plt.scatter(subset['Frequency'], subset['Monetary'], label=seg,
                alpha=0.6, color=color, s=40)
plt.xlabel('Frequency')
plt.ylabel('Monetary Value')
plt.title('Customer Segments by Frequency vs Monetary (RFM Score-Based)')
plt.legend(title='Segment')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Define 90-day prediction window
absolute_max_date = df['InvoiceDate'].max()
prediction_cutoff_date = absolute_max_date - pd.Timedelta(days=90)

print(f"Absolute Max Invoice Date: {absolute_max_date}")
print(f"Prediction Cutoff Date:    {prediction_cutoff_date}")

historical_df = df[df['InvoiceDate'] <= prediction_cutoff_date]
future_df     = df[df['InvoiceDate'] >  prediction_cutoff_date]

print(f"Historical rows: {historical_df.shape[0]}")
print(f"Future rows:     {future_df.shape[0]}")
```

```python
# Recompute RFM on historical period
rfm_hist = historical_df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (prediction_cutoff_date - x.max()).days,
    'InvoiceNo':   'nunique',
    'TotalPrice':  'sum'
})
rfm_hist.columns = ['Recency', 'Frequency', 'Monetary']
rfm_hist = rfm_hist[rfm_hist['Monetary'] > 0]
display(rfm_hist.head())
```

```python
# Calculate actual 90-day CLV (target variable)
prediction_window_end_date = prediction_cutoff_date + pd.Timedelta(days=90)
future_90_day_df = future_df[
    (future_df['InvoiceDate'] > prediction_cutoff_date) &
    (future_df['InvoiceDate'] <= prediction_window_end_date)
]
future_clv = future_90_day_df.groupby('CustomerID')['TotalPrice'].sum().rename('90DayCLV')
print(f"Customers with future transactions: {len(future_clv)}")
display(future_clv.head())
```

```python
# Assign RFM scores to historical RFM data
rfm_hist['R_Score'] = pd.qcut(rfm_hist['Recency'], q=3, labels=[3, 2, 1]).astype(int)
rfm_hist['F_Score'] = pd.qcut(rfm_hist['Frequency'].rank(method='first'), q=3, labels=[1, 2, 3]).astype(int)
rfm_hist['M_Score'] = pd.qcut(rfm_hist['Monetary'], q=3, labels=[1, 2, 3]).astype(int)
rfm_hist['RFM_Score'] = rfm_hist['R_Score'] + rfm_hist['F_Score'] + rfm_hist['M_Score']

rfm_hist['Segment'] = rfm_hist['RFM_Score'].apply(lambda s: 
    'High Value' if s >= 7 else ('Medium Value' if s >= 4 else 'Low Value'))

print("Segment distribution (historical):")
print(rfm_hist['Segment'].value_counts())
```

```python
# Merge historical RFM + segments with 90-day CLV target
clv_data = rfm_hist.merge(future_clv, on='CustomerID', how='left').fillna(0)
print(f"CLV data shape: {clv_data.shape}")
print(f"Columns: {clv_data.columns.tolist()}")
display(clv_data.head())
```

```python
models = {}
results = {}

print("Available Segments:", clv_data['Segment'].unique())

for segment in ['High Value', 'Medium Value', 'Low Value']:
    seg_data = clv_data[clv_data['Segment'] == segment]
    
    if len(seg_data) > 5:
        X_c = seg_data[['Recency', 'Frequency', 'Monetary']]
        y_c = seg_data['90DayCLV']
        
        X_train, X_test, y_train, y_test = train_test_split(
            X_c, y_c, test_size=0.2, random_state=42
        )
        
        model = Ridge(alpha=1.0)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        mse = mean_squared_error(y_test, y_pred)
        r2  = r2_score(y_test, y_pred)
        
        models[segment]  = model
        results[segment] = {'mse': mse, 'r2': r2}
        print(f"✓ Model trained for: {segment} | MSE={mse:.2f} | R²={r2:.4f}")
    else:
        print(f"⚠ Not enough data for segment: {segment}")

print(f"\n✅ Successfully trained {len(models)} models")
```

```python
fig, axes = plt.subplots(1, len(models), figsize=(6*len(models), 5))
if len(models) == 1:
    axes = [axes]

for ax, (segment, model) in zip(axes, models.items()):
    seg_data = clv_data[clv_data['Segment'] == segment]
    X_c = seg_data[['Recency', 'Frequency', 'Monetary']]
    y_c = seg_data['90DayCLV']
    _, X_test, _, y_test = train_test_split(X_c, y_c, test_size=0.2, random_state=42)
    y_pred = model.predict(X_test)
    ax.scatter(y_test, y_pred, alpha=0.6)
    ax.set_title(f'{segment}\nR²={results[segment]["r2"]:.4f}')
    ax.set_xlabel('Actual CLV')
    ax.set_ylabel('Predicted CLV')
    ax.grid(True, alpha=0.3)

plt.suptitle('Actual vs Predicted 90-Day CLV per Segment', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
joblib.dump(models, 'clv_regression_models.joblib')
print(f"✅ Models saved as 'clv_regression_models.joblib'")
print(f"   Segments: {list(models.keys())}")
```
